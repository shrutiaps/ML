```python
#Importing libraries
from sklearn.datasets import load_iris
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import BaggingClassifier
from sklearn.metrics import accuracy_score


#Load the iris dataset
data = load_iris()

#Create a dataframe for the iris dataset
df = pd.DataFrame(data.data, columns = data.feature_names)

#Adding the target column to the dataframe
df['Species'] = data.target

#Displaying the dataframe
df
```

```python
#Displaying information about the dataset such as number of columns and data types
df.info()
```

```python
#Displaying the statistical summary of the dataset
df.describe()
```

```python
#Defining target and features
x = df[data.feature_names]
y = df['Species']


#Splitting the data
xtrain, xtest, ytrain, ytest = train_test_split(x,y,test_size = 0.25, random_state = 47)
print(xtrain.shape, xtest.shape, ytrain.shape, ytest.shape)
```

```python
#Fitting the data on the training data
dt = DecisionTreeClassifier()
dt.fit(xtrain, ytrain)
```

```python
#Obtaining predictions for the test data
predictiondtc = dt.predict(xtest)
predictiondtc
```

```python
rfc = RandomForestClassifier()
rfc.fit(xtrain, ytrain)
```

```python
#Making predictions with random forest classifier
predictionrf = rfc.predict(xtest)

#Displaying the predictions of the Random Forest Classifier
predictionrf
```

```python
model = BaggingClassifier()
model.fit(xtrain, ytrain)
```

```python

predictionb = model.predict(xtest)
predictionb
```

```python

print("Accuracy Score of DTC: ", accuracy_score(predictiondtc, ytest))
print("Accuracy Score of RFC: ", accuracy_score(predictionrf, ytest))
print("Accuracy Score of Bagging Classifier: ", accuracy_score(predictionb, ytest))
```
